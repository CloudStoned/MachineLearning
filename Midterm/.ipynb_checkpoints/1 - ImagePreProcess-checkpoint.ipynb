```python
import cv2
import matplotlib.pyplot as plt
import numpy as np 
```

### Gaussian Blur

```python
# Apply Gaussian blur filter
gaussian_filtered = cv2.GaussianBlur(noisy_image, (9, 9), 1.5)

# Apply bilateral filtering
bilateral_filtered = cv2.bilateralFilter(gaussian_filtered, d=9, sigmaColor=75, sigmaSpace=75)

# Apply non-local means denoising
nlmeans_denoised = cv2.fastNlMeansDenoisingColored(bilateral_filtered, None, h=10, hColor=10, templateWindowSize=7, searchWindowSize=21)

# Display the results
plt.figure(figsize=(15, 5))

plt.subplot(1, 4, 1)
plt.imshow(cv2.cvtColor(noisy_image, cv2.COLOR_BGR2RGB))
plt.title('Noisy Image')
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(cv2.cvtColor(gaussian_filtered, cv2.COLOR_BGR2RGB))
plt.title('Gaussian Filtered')
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(cv2.cvtColor(bilateral_filtered, cv2.COLOR_BGR2RGB))
plt.title('Bilateral Filtered')
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(cv2.cvtColor(nlmeans_denoised, cv2.COLOR_BGR2RGB))
plt.title('Non-Local Means Denoised')
plt.axis('off')

plt.show()
```

### Contrast Stretching

```python
xray_img = cv2.imread('data/Images/xray.jpg')

plt.imshow(cv2.cvtColor(xray_img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
img = cv2.imread('data/Images/xray.jpg', cv2.IMREAD_GRAYSCALE)

# Find min and max values in the image
min_val = np.min(img)
max_val = np.max(img)

# Display the grayscale image using OpenCV
cv2.imshow('Grayscale', xray_img)
cv2.waitKey(0)
cv2.destroyAllWindows()

# Display the grayscale image using matplotlib
plt.imshow(img, cmap='gray')
plt.axis('off')
plt.show()
```

### Normalization

```python
norm_img = cv2.normalize(xray_img, None, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_32F)
cv2.imshow('Normalized Image', norm_img)
plt.imshow(cv2.cvtColor(norm_img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

### Resize Image

```python
resized_img = cv2.resize(xray_img, (120, 210))
cv2.imshow('Resized Image', resized_img)
plt.imshow(cv2.cvtColor(resized_img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show
```

### Canny Image

```python
# Convert the image to grayscale
gray_img = cv2.cvtColor(xray_img, cv2.COLOR_BGR2GRAY)

# Detect edges
edges = cv2.Canny(gray_img, threshold1=10, threshold2=20)


# Display the edges using Matplotlib
plt.imshow(edges, cmap='gray')  # Use cmap='gray' for grayscale images
plt.axis('off')
plt.show()
```

### Flip Image

```python
import cv2
import matplotlib.pyplot as plt

# Assuming xray_img is already loaded properly

# Flip the image horizontally
flipped_image = cv2.flip(xray_img, 1)


# Display the flipped image using Matplotlib
plt.imshow(cv2.cvtColor(flipped_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
