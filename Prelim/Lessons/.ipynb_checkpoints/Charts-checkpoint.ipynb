```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
```

```python
house_sizes = np.array([1000, 1500, 2000, 2500, 3000]).reshape(-1,1)
sale_prices = np.array([200000,300000,400000,500000,600000])
```

```python
plt.figure(figsize=(8,6))
plt.scatter(house_sizes, sale_prices, color='red',label='Data Points')
plt.title('House sizes vs Sale Price')
plt.xlabel('Size')
plt.ylabel('Sale Price')
plt.legend()
```

```python
model = LinearRegression()
model.fit(house_sizes, sale_prices)

plt.plot(house_sizes, model.predict(house_sizes), color='blue', label='linear regression')
plt.legend()
plt.show()
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data"
column_names = ['Class', 'Alcohol', 'Malic acid', 'Ash', 'Alcalinity of ash', 'Magnesium', 
                'Total phenols', 'Flavanoids', 'Nonflavanoid phenols', 'Proanthocyanins', 
                'Color intensity', 'Hue', 'OD280/OD315 of diluted wines', 'Proline']
data = pd.read_csv(url, names=column_names)
```

### Histogram

```python
# Kung anong column ang gusto iexamine
column_of_interest =  'Alcohol'

plt.figure(figsize = (8,6))
plt.hist(data[column_of_interest], bins = 20, color = 'cyan', edgecolor = 'black', alpha=0.7)
plt.title('Histogram')
plt.xlabel(column_of_interest)
plt.ylabel('Frequency')
plt.show()
```

### Line Chart

```python
x = np.linspace(0, 10, 100)
y = 2 * x + 3 + np.random.normal(0, 2, 100)
```

```python
plt.figure(figsize = (8,6))
plt.plot(x,y, label='Data')
plt.title('x-axis')
plt.xlabel('y-axis')
plt.ylabel('Frequency')
plt.title('Identifying a trend with data visualization')
plt.legend()
plt.show()
```

```python
flights_data = pd.read_csv("https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv")
flights_data1 = flights_data.iloc[:30,:]
```

```python
flights_data1.tail()
```

```python
plt.figure(figsize=(10, 6)) 
plt.plot(flights_data1['Month'], flights_data1['Passengers'], marker='o', linestyle='-')
plt.title('Passenger Traffic Over Time', fontsize=16) 
plt.xlabel('Month', fontsize=12) 
plt.ylabel('Passengers', fontsize=12) 
plt.grid(True, linestyle='--', alpha=0.7)  
plt.legend(['Passenger Count'], loc='upper left', fontsize=12)  
plt.xticks(rotation=55, ha='right')  
plt.tight_layout()  
plt.show()
```
