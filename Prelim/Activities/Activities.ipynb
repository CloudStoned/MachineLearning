### Activity 3: Data Collection and Inspection 

```python
df_wine = pd.read_csv('data/Wine/winequality-red.csv', delimiter=';', quoting = 1)
df_wine
```

### Activity 4: Handling Missing Values and Duplicates

```python
missing_values_wine = df_wine.dropna()
missing_values_wine
```

```python
duplicate_rows = df_wine[df_wine.duplicated(keep=False)]
duplicate_rows
```

```python
clean_wine = df_wine.drop_duplicates()
clean_wine
```

```python
rows_before_removal = df_wine.shape[0]
df_no_duplicates = df_wine.drop_duplicates()
rows_after_removal = df_no_duplicates.shape[0]
rows_removed_count = rows_before_removal - rows_after_removal

print("Number of Rows Before Removal:", rows_before_removal)
print("Number of Rows After Removal:", rows_after_removal)
print("Number of Rows Removed:", rows_removed_count)
```

###  Activity 5: Feature Engineering - Creating New Features

```python
sum_of_fixed_acidity = clean_wine['fixed acidity'].sum()
sum_of_volatile_acidity = clean_wine['volatile acidity'].sum()

total_acidity = sum_of_fixed_acidity + sum_of_volatile_acidity

print('Sum of Fixed Acidity: ', sum_of_fixed_acidity)
print('Sum of Volatile Acidity: ', sum_of_volatile_acidity)
print('Total Acidity: ', total_acidity)
```

```python
clean_wine.loc[:, 'Total Acidity'] = clean_wine['fixed acidity'] + clean_wine['volatile acidity']
clean_wine.loc[:, 'Wine Quality'] = clean_wine['quality'].apply(lambda x: '1' if x >= 7 else '0')
clean_wine.head(15)
```
